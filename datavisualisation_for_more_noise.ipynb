```python
import glob
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
import jax.numpy as jnp

gloober = glob.glob("grid_search11/*")

figs, axs = plt.subplots(1,2,figsize=(9,5))

learning_rate = [0.01, 0.001]

for idx, glob in enumerate(gloober):
    with open(glob, "rb") as f:
        results = pkl.load(f)
        print(results.keys())
        axs[idx].plot(results["Accuracy Measurements"]["Training accuracy"], label="Train")
        axs[idx].plot(results["Accuracy Measurements"]["Testing accuracy"], label="Test")
        axs[idx].grid()
        axs[idx].set_title(f"Learning Rate : {learning_rate[idx]}")
        axs[idx].legend()

plt.savefig("20kHzwiththenoise.pdf")
```
